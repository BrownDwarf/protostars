## 07- Probabilistic Graphical Models
### I. PGM for veiling

This notebook is adapted from the notebook available in the Starfish repo.  The figure is modeled after an identical one in Czekala et al. 2015.

```python
from matplotlib import rc
rc("font", family="serif", size=10)
rc("text", usetex=True)

import matplotlib.pyplot as plt
import daft
```

### Original figure
Modified to have a larger *y*-extent

```python
# Instantiate the PGM.
pgm = daft.PGM([4.0, 3.0], origin=[0.3, 0.3])

# Parameters.
pgm.add_node(daft.Node("star", r"$\mathbf{\theta}_\star$", 0.9, 2))
pgm.add_node(daft.Node("w", r"$\mathbf{w}$", 1.8, 2, plot_params={"fill":True, "fc":"w"}))

pgm.add_node(daft.Node("model", r"$\mathbf{\mathsf{M}}$", 2.7, 2, plot_params={"fill":True, "fc":"w"}))

pgm.add_node(daft.Node("eig", r"$\mathbf{\Xi}$", 2.0, 1, fixed=True))
pgm.add_node(daft.Node("obs", r"$\mathbf{\theta}_\textrm{obs}$", 2.55, 1))
pgm.add_node(daft.Node("phi", r"$\phi_{\mathsf{P}}$", 3.1, 1))

pgm.add_node(daft.Node("data", r"$\mathbf{\mathsf{D}}$", 3.7, 2, observed=True))
pgm.add_node(daft.Node("nuis", r"$\phi_\mathsf{C}$", 3.7, 1))



# Add in the edges.
pgm.add_edge("star", "w")
pgm.add_edge("w", "model")
pgm.add_edge("model", "data")



pgm.add_edge("eig", "model")
pgm.add_edge("obs", "model")
pgm.add_edge("phi", "model")
pgm.add_edge("nuis", "data")

pgm.render()
pgm.figure.savefig("../plots/PGM_original.pdf")
```

### Add a veiling node

```python
# Instantiate the PGM.
pgm = daft.PGM([4.0, 3.0], origin=[0.3, 0.3])

# Parameters.
pgm.add_node(daft.Node("star", r"$\mathbf{\theta}_\star$", 0.9, 2))
pgm.add_node(daft.Node("w", r"$\mathbf{w}$", 1.8, 2, plot_params={"fill":True, "fc":"w"}))

pgm.add_node(daft.Node("model", r"$\mathbf{\mathsf{M}}$", 2.7, 2, plot_params={"fill":True, "fc":"w"}))

pgm.add_node(daft.Node("eig", r"$\mathbf{\Xi}$", 2.0, 1, fixed=True))
pgm.add_node(daft.Node("obs", r"$\mathbf{\theta}_\textrm{obs}$", 2.55, 1))
pgm.add_node(daft.Node("phi", r"$\phi_{\mathsf{P}}$", 3.1, 1))

pgm.add_node(daft.Node("data", r"$\mathbf{\mathsf{D}}$", 3.7, 2, observed=True))
pgm.add_node(daft.Node("nuis", r"$\phi_\mathsf{C}$", 3.7, 1))

#Veiling
pgm.add_node(daft.Node("veil", r"$[T_{d}, \Omega_{d}]$", 1.8, 2.7, aspect=1.45, plot_params={"fill":True, "fc":"w"}))


# Add in the edges.
pgm.add_edge("star", "w")
pgm.add_edge("w", "model")
pgm.add_edge("model", "data")
pgm.add_edge("veil", "model")


pgm.add_edge("eig", "model")
pgm.add_edge("obs", "model")
pgm.add_edge("phi", "model")
pgm.add_edge("nuis", "data")

pgm.render()
#pgm.figure.savefig("../plots/PGM_veiling.pdf")
pgm.figure.savefig("../plots/PGM_veiling.png", dpi=300)
```
